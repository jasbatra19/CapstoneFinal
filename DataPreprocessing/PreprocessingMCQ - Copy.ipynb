```python
import pandas as pd
```

```python
df = pd.read_csv("../QuestionGenerator/mcqAnswer/questionAnswer.csv")
df
```

### Preprocessing of questions

```python
questions=df['questions']
question=[]
for data in questions:
    entry=data.replace("_","")
    entry=entry.strip().split(".")
    question.append(entry[1])
df['question']=question

```

## Preprocessing of options for mcqs

```python
answers=df['answers']
optionA=[]
optionB=[]
optionC=[]
optionD=[]
answer=[]
explanation=[]
for data in answers:
    entry = data.replace('A.', "   ").replace('B.', "   ").replace('C.', "   ").replace("D.", "   ").replace("View Answer", "   ").replace("\\n", "").replace("'",'').replace("\"",'').replace("Ans : ","   ").replace("Explanation:","   ")
    entry = entry.split(",")
    
    # Further split each element based on spaces
    split_entries = [part.strip() for part in entry[0].split('   ') if part.strip()]
    # print(split_entries)
    while len(split_entries)<5:
        split_entries.append("")
    optionA.append(split_entries[1])
    optionB.append(split_entries[2])
    optionC.append(split_entries[3])
    optionD.append(split_entries[4])

    split_answers=[part.strip() for part in entry[1].split("   ") if part.strip()]
    answer.append(split_answers[0])
    while len(split_answers)<2:
        split_answers.append("")
    explanation.append(split_answers[1])


    

          
    
df['A']=optionA
df['B']=optionB
df['C']=optionC
df['D']=optionD
df['answer']=answer
df['explanation']=explanation



df
```

```python
df = df.drop(columns=['questions','answers'])
```
